# Top performing models so far

```python
# 35.5 seconds with 55% accuracy.

batch_size = 100
num_epochs = 8


class Model(torch.nn.Module):
    def __init__(self, num_channels: int, num_classes: int) -> None:
        super().__init__()
        """
        Model
        """

        # (32, 32, 3)
        self.conv1 = torch.nn.Conv2d(
            in_channels=num_channels,
            out_channels=32,
            kernel_size=3,
            padding=1,
            stride=1,
        )
        # (32, 32, 8)
        self.relu1 = torch.nn.ReLU()
        self.maxpool1 = torch.nn.MaxPool2d(kernel_size=2, stride=2)
        # (16, 16, 8)

        # (16, 16, 8)
        self.conv2 = torch.nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3)
        # (14, 14, 16)
        self.relu2 = torch.nn.ReLU()
        self.maxpool2 = torch.nn.MaxPool2d(kernel_size=2, stride=2)
        # (7, 7, 16)
        self.fc1 = torch.nn.Linear(7 * 7 * 64, 30)
        self.relu3 = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(30, num_classes)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        Forward function
        """

        x = self.conv1(x)
        x = self.relu1(x)
        x = self.maxpool1(x)

        x = self.conv2(x)
        x = self.relu2(x)
        x = self.maxpool2(x)

        x = torch.flatten(x, start_dim=1)

        x = self.fc1(x)
        x = self.fc2(x)
        return x
```
